2. Poverty Scale (Headcount) vs. Severity (Intensity)What the data reveals: There is a strong linear correlation between the percentage of people who are poor ($H$) and the average share of deprivations they experience ($A$). As headcount ratio increases, the intensity of poverty also deepens.Why the insights matter: This indicates that poverty is not just about "how many" people are poor, but also about the "depth" of their struggle. The $MPI = H \times A$ formula captures this compounding effect.Real-world context: In regions with high headcount and high intensity, people aren't just lacking one thing (like income); they likely lack clean water, electricity, schooling, and nutrition simultaneously, creating a "poverty trap" that is harder to escape.

```python
top_10_regions = data.nlargest(10, 'MPI Regional')
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_regions, x='MPI Regional', y='Sub-national region', hue='Sub-national region', palette='inferno', legend=False)

plt.title('Top 10 Most Impoverished Sub-national Regions', fontsize=25)
plt.xlabel('Regional MPI')
plt.ylabel('Sub-national Region')
plt.tight_layout()
plt.savefig('3_top_10_regions.png')
plt.show()
```

3. Top 10 Most Impoverished Sub-national RegionsWhat the data reveals: The most impoverished regions globally are concentrated in a few countries, particularly within Sub-Saharan Africa (e.g., regions in South Sudan, Chad, and Ethiopia). These regions have MPI scores exceeding $0.6$.Why the insights matter: These are the "global hotspots" of deprivation. Targeting these specific sub-national entities is more effective than broad national-level interventions.Real-world context: A region like Wadi Fira in Chad or Warrap in South Sudan faces levels of deprivation that are exponentially higher than the global average, necessitating specialized humanitarian aid and infrastructure development.4. Intra-country MPI Variation

```python
data = pd.read_csv('MPI_subnational.csv').dropna(subset=['Intensity of deprivation Regional'])

top_countries = data['Country'].value_counts().nlargest(10).index
subset_countries = data[data['Country'].isin(top_countries)]
avg_mpi_country = subset_countries.groupby('Country')['MPI Regional'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(12, 8))
sns.barplot(data=avg_mpi_country, x='Country', y='MPI Regional', hue='Country', palette='coolwarm', legend=False)
plt.title('Average Regional MPI (Top 10 Most Represented Countries)', fontsize=25)
plt.xticks(rotation=0)
plt.ylabel('Average Regional MPI')
plt.xlabel('Country')
plt.tight_layout()
plt.savefig('1_avg_mpi_by_world_region.png')
plt.show()
```

4. Intra-country MPI VariationWhat the data reveals: Within countries like Nigeria or Ethiopia, there is a massive range in regional MPI. Some regions may have low poverty (comparable to Latin America), while others in the same country face extreme deprivation.Why the insights matter: It highlights internal inequality. A "low" national MPI can be misleading if a country has one wealthy capital region and several severely impoverished rural provinces.Real-world context: This justifies the need for decentralized governance. Policy solutions for a capital city cannot be applied to a remote province where $80\%$ of the population lacks basic sanitation.

```python
avg_headcount = data.groupby('World region')['Headcount Ratio Regional'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Headcount Ratio Regional', y='World region', data=avg_headcount,hue='World region',  palette='Dark2',legend=False)
plt.title('Average Headcount Ratio by World Region', fontsize=25)
plt.xlabel('Average % of Population in Poverty')
plt.ylabel('World Region')
plt.tight_layout()
plt.savefig('5_avg_headcount_region.png')
plt.show()
```

5. Average Headcount Ratio by World RegionWhat the data reveals: In Sub-Saharan Africa, the average regional headcount ratio is over $60\%$, whereas in Europe and Central Asia, it is significantly lower. South Asia also shows a high average headcount, exceeding $40\%$.Why the insights matter: This clarifies the sheer scale of the population affected. It shows that in certain parts of the world, being multidimensionally poor is the "norm" rather than the exception.Real-world context: High headcount ratios place immense pressure on public services (health and education) as the majority of the population requires subsidized support to meet basic needs.

```python
plt.figure(figsize=(12, 8))
sns.histplot(data['Intensity of deprivation Regional'], bins=30, kde=True, color='#003366')
plt.title('Global Distribution of Poverty Intensity', fontsize=25)
plt.xlabel('Intensity of Deprivation (%)')
plt.ylabel('Frequency (Number of Regions)')
plt.tight_layout()
plt.savefig('6_intensity_distribution.png')
plt.show()
```

6. Global Distribution of Poverty IntensityWhat the data reveals: The intensity of deprivation ($\%$) across most impoverished regions clusters between $40\%$ and $55\%$. It is rare to see intensity below $33\%$ (the threshold for being "poor") or above $75\%$.Why the insights matter: This identifies a "poverty floor." Once a region is classified as poor, the average person is typically deprived in at least half of the weighted indicators.Real-world context: This suggests that poverty is rarely a single-issue problem. For the average "poor" individual in these regions, policy must address a bundle of needs (e.g., health + education + assets) because the intensity of their deprivation is structurally deep.
